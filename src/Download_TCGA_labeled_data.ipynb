```python
import os
import pandas as pd

# assign directory
directory = '/content/data'

all_dir = [os.path.join(directory, filename) for filename in os.listdir(directory) if os.path.isdir(os.path.join(directory, filename))]
all_files = []
for folder in all_dir:
  for filename in os.listdir(folder):
      f = os.path.join(folder, filename)
      if os.path.isfile(f):
        all_files.append(f)

all_files = [file for file in all_files if file.endswith('.tsv')]
```

```python
df = pd.read_csv(all_files[0], sep='\t', skiprows=[0, 2, 3, 4, 5], usecols=[0, 6])
general_df = df.rename(columns={df.columns[1]: all_files[0].split('/')[3]})

for f in all_files[1:]:
  df = pd.read_csv(f, sep='\t', skiprows=[0, 2, 3, 4, 5], usecols=[0, 6])
  df = df.rename(columns={df.columns[1]: f.split('/')[3]})
  general_df = pd.merge(general_df, df, on='gene_id', how='inner')


general_df
```

```python
general_df.to_csv('kidney.tsv', index=False, sep="\t")
```
